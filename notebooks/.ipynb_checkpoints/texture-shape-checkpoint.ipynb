## UAP Visualizations: Texture vs. Shape-biased Models
Visualizations of some targeted UAPs on ImageNet for the texture and shape-biased ResNet50 models.

```python
import json
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
dir_uap = '../uaps/imagenet/'
all_models = ['resnet50', 'resnet50_SIN',  'resnet50_SIN-IN']
shorthand = ['ResNet50 (IN)', 'ResNet50 (SIN)', 'ResNet50 (SIN+IN)']

with open('../imgnet_labels.json', 'r') as j:
     dict_labels = json.loads(j.read())
```

```python
def display_example(y_tgt):
    plt.figure(figsize = (15, 5))
    plt.subplots_adjust(wspace = 0.04)
    plt.suptitle('Target: %i (%s)' % (y_tgt, dict_labels[str(y_tgt)]), fontsize = 28, y = 1.05)

    for test_model in all_models:
        idx = all_models.index(test_model)
        plt.subplot(1, len(all_models), idx + 1)
        p = torch.load(dir_uap + 'sgd-tgt%i-%s-eps10.pth' % (y_tgt, test_model))
        plt.title(shorthand[idx], fontsize = 24)
        plt.imshow((((p.data / torch.max(p.data)) + 1) / 2).permute(1, 2, 0))
        plt.tick_params(axis='both', which='both', left=False, bottom=False,
                        top=False, labelbottom=False, labelleft=False)
```

```python
for y_target in [611, 854, 859]:
    display_example(y_target)
```
